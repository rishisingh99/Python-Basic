**Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.**
Answer:

Statistics is broadly divided into descriptive statistics and inferential statistics, based on how data is used and interpreted.

**Descriptive Statistics**

Descriptive statistics deal with summarizing and organizing data so that it can be easily understood. It does not involve making predictions or generalizations beyond the given data.

Key tools used:

Mean, Median, Mode

Range, Variance, Standard Deviation

Tables, Charts, Graphs (Bar chart, Pie chart, Histogram)

Example:
If a class of 30 students has an average score of 75 marks, descriptive statistics simply describe this result without drawing conclusions about other classes.

**Inferential Statistics**

Inferential statistics involve drawing conclusions or making predictions about a population based on a sample of data.

Key tools used:

Hypothesis testing

Confidence intervals

Correlation and regression

Probability distributions

Example:
If we survey 50 customers and conclude that 80% of all customers are satisfied, this is inferential statistics because we are generalizing from a sample to a population.

Difference Summary
Descriptive Statistics	Inferential Statistics
Describes data	Predicts or generalizes
Uses entire dataset	Uses sample data
No prediction	Involves estimation
Example: Average marks	Example: Predicting future results


**Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.**
Answer:
Sampling

Sampling is the process of selecting a subset of individuals or observations from a population to represent the entire population.

Sampling is used because:

Studying the entire population is costly and time-consuming

Sampling saves time and resources

Accurate results can still be obtained

**Random Sampling**

In random sampling, every individual has an equal chance of being selected.

Characteristics:

Simple and unbiased

Easy to implement

May not represent subgroups properly

Example:
Randomly selecting 100 students from a college list.

**Stratified Sampling**

In stratified sampling, the population is divided into subgroups (strata), and samples are taken from each group.

Characteristics:

Ensures representation of all subgroups

More accurate results

Slightly complex to implement

Example:
Dividing employees into departments (HR, IT, Sales) and selecting samples from each department.

Comparison Table
Random Sampling	Stratified Sampling
Equal chance for all	Equal representation
Simple method	More accurate
Can miss subgroups	Covers all groups

**Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**
Answer:
Skewness

Skewness measures the asymmetry of a distribution.

Positive skew (Right skew): Long tail on the right

Negative skew (Left skew): Long tail on the left

Zero skew: Symmetrical distribution

Positive Skew

A positive skew indicates:

Most values are low

Few extremely high values

Mean > Median > Mode

Example: Income distribution (few very high incomes)

Kurtosis

Kurtosis measures the peakedness or flatness of a distribution.

High kurtosis: Sharp peak

Low kurtosis: Flat distribution





```python
#Question 5: Implement a Python program to compute mean, median, and mode.
#Answer:

#Python Code:

import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)

```

```python
#Question 6: Compute covariance and correlation coefficient.
#Answer:

#Python Code:

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0,1]
correlation = np.corrcoef(list_x, list_y)[0,1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```

```python
#Question 7: Draw a boxplot and identify outliers.
#Answer:

#Python Code:

import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.show()


#Result Explanation:

#Value 35 appears as an outlier

#Data is slightly right-skewed
```

```python
#Question 8: Relationship between advertising spend and sales.
#Answer:
#Explanation

#Covariance shows direction of relationship

#Correlation shows strength of relationship

#Positive values indicate sales increase with ad spend

#Python Code:

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

correlation = np.corrcoef(advertising_spend, daily_sales)[0,1]
print("Correlation:", correlation)





#Conclusion:
#Very strong positive relationship.

```
